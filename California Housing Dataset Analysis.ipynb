1. Longitude - Longitude is a geographic coordinate that specifies the east–west position of a point

3. house_median_age - It is the median age of house
4. total_rooms - Total number of rooms in a block

9. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
#creating a user define function to categorize bedrooms size based on number of bedrooms
def bedroom_size(total_bedrooms):
  if total_bedrooms <= 10:
    return("small")
  elif 11 <= total_bedrooms <1000:
    return("medium")
  else:
    return("large")
#using map function to apply bedroom_size(user defined function) to total_bedrooms featur of housing and creating a new coloumn total_bedroom_size to add the category information
housing["total_bedroom_size"]=housing["total_bedrooms"].map(bedroom_size)

#printing the dataframe
housing.head()
```

**Description**
In the above task I createda user defined function which categorized the households based on the number of bedrooms.
